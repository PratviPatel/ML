```python
iris = pd.read_csv('iris.csv')
```

```python
iris
```

```python
x = iris[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
y = iris['species']
```

```python
x
```

```python
y
```

```python
# y has string values , so convert into numbers

from sklearn.preprocessing import LabelEncoder
```

```python
lbly = LabelEncoder()
```

```python
y = lbly.fit_transform(y)
```

```python
y
```

```python
#split data to test and train

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
#Logistic Regression

from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
y_pred = logmodel.predict(x_test)
```

```python
#confusion metrix

from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
logmodel.score(x_test,y_test)
```

```python
#KNeighbours Classifier

from sklearn.neighbors import KNeighborsClassifier
```

```python
knn= KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred_knn = knn.predict(x_test)
```

```python
confusion_matrix(y_test, y_pred_knn)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred_knn,y_test)
```

```python
#Navie Bayes

from sklearn.naive_bayes import GaussianNB
```

```python
bayes = GaussianNB()
```

```python
bayes.fit(x_train,y_train)
```

```python
y_pred_bayes = bayes.predict(x_test)
```

```python
accuracy_score(y_pred_bayes,y_test)
```

```python
bayes.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
class_svc = SVC(kernel='linear')
```

```python
class_svc.fit(x_train,y_train)
```

```python
y_pred_svc = class_svc.predict(x_test)
```

```python
confusion_matrix(y_pred_svc,y_test)
```

```python
accuracy_score(y_pred_svc,y_test)
```
